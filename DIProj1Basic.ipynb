```python
# -*- coding: utf-8 -*-
"""
Created on Mon Jul 29 12:26:55 2019

@author: rahul
"""

import csv
import pandas as pd
import os
from datetime import datetime
import matplotlib as plt
%matplotlib inline

current_directory2 = "D:\Summer\DataIncubator\Datasets"
dir3 = "temp_datalab_records_job_listings"
#os.getcwd()
fileName2 = "temp_datalab_records_job_listings.csv"
completePath2 = os.path.join(current_directory2,dir3,fileName2)


jobsDF = pd.read_csv(completePath2,nrows=1000000)






```

```python

# 
def dateToYear(data):
    year = data[:4]
#    print(year)
    return int(year)


def dateToMonth(data):
    month = data[5:7]
#    print(year)
    return int(month)


import warnings
warnings.filterwarnings('ignore')
```

```python
jobsDF.columns
jobsDF1 = jobsDF.dropna(subset=['posted_date','country'])

jobsDF1["Year"] =  jobsDF1["posted_date"].apply(lambda x: dateToYear(x))

jobsDF1_2017 = jobsDF1[jobsDF1["Year"] == 2018] 
gb_country = jobsDF1_2017.groupby("country").count()

top10 = gb_country["dataset_id"].nlargest(10)

top10Label = top10.index.tolist()
#top10Count = top10.dataset_id.tolist()

plot1 = top10.plot.bar()
plot1.set_title("Number of jobs in 2018 in 10 countries having highest number of jobs")
```

```python
jobsDFUSA = jobsDF[jobsDF["country"] == "USA"]
jobsDFUSA = jobsDFUSA.dropna(subset=['posted_date'])
jobsDFUSA["Year"] =  jobsDFUSA["posted_date"].apply(lambda x: dateToYear(x))
jobsDFUSA["Month"] =  jobsDFUSA["posted_date"].apply(lambda x: dateToMonth(x))


gb_year_USA = jobsDFUSA.groupby("Year").count()
gb_month_USA_2018 = jobsDFUSA[jobsDFUSA["Year"] == 2018].groupby("Month").count()
gb_month_USA_2017 = jobsDFUSA[jobsDFUSA["Year"] == 2017].groupby("Month").count()
gb_month_USA_2016 = jobsDFUSA[jobsDFUSA["Year"] == 2016].groupby("Month").count()
gb_month_USA_2015 = jobsDFUSA[jobsDFUSA["Year"] == 2015].groupby("Month").count()

# gb_month_USA_2015["dataset_id"].plot.bar()

# gb_month_USA_2016["dataset_id"].plot.bar()


plot2 = gb_month_USA_2017["dataset_id"].plot.bar()

plot2.set_title("Number of jobs in USA for different months in 2017")
# gb_month_USA_2018["dataset_id"].plot.bar()
```
